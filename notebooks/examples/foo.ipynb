```python
from icecap.infrastructure.memory_manager import get_memory_manager
from icecap.infrastructure.process import get_wow_process_id
from icecap.infrastructure.driver import GameDriver
from icecap.infrastructure.repository import PlayerRepository, UnitRepository
import numpy as np
import pyqtgraph as pg
from pyqtgraph.Qt import QtCore
import time

memory_manager = get_memory_manager(get_wow_process_id())
driver = GameDriver(memory_manager)
player_repository = PlayerRepository(driver)

local_player = player_repository.get_local_player()
initial_player_position = local_player.position
print(f"Initial player position: {initial_player_position}")

# Set up the plot
app = pg.mkQApp("Players")
win = pg.GraphicsLayoutWidget(show=True, title="Player Positions")
win.resize(800, 600)
plot = win.addPlot(title="Player Positions")
plot.setLabel('left', 'Y Coordinate')
plot.setLabel('bottom', 'X Coordinate')
plot.addLegend()
#plot.showGrid(x=True, y=True)

# Create scatter plots for player (blue) and units (orange)
player_scatter = pg.ScatterPlotItem(size=15, pen=pg.mkPen(None), brush=pg.mkBrush(30, 114, 255, 255))
enemy_scatter = pg.ScatterPlotItem(size=10, pen=pg.mkPen(None), brush=pg.mkBrush(255, 0, 0, 200))
friendly_scatter = pg.ScatterPlotItem(size=10, pen=pg.mkPen(None), brush=pg.mkBrush(0, 255, 0, 200))
plot.addItem(player_scatter)
plot.addItem(enemy_scatter)
plot.addItem(friendly_scatter)

# Initial player position
player_x = [initial_player_position.x]
player_y = [initial_player_position.y]
player_scatter.setData(player_x, player_y)

# Initial unit positions

# Initial player and other players positions
all_players = list(player_repository.yield_players())

# Separate initial players into enemy and friendly groups
enemy_positions_x = []
enemy_positions_y = []
friendly_positions_x = []
friendly_positions_y = []

for player in all_players:
    if player.guid != local_player.guid and hasattr(player, 'position'):
        if local_player.is_enemy(player):
            enemy_positions_x.append(player.position.x)
            enemy_positions_y.append(player.position.y)
        else:
            friendly_positions_x.append(player.position.x)
            friendly_positions_y.append(player.position.y)

# Set initial data for enemy and friendly scatter plots
enemy_scatter.setData(enemy_positions_x, enemy_positions_y)
friendly_scatter.setData(friendly_positions_x, friendly_positions_y)


def update():
    # Refresh player and get new position
    refreshed_player = player_repository.refresh_player(local_player)
    player_position = refreshed_player.position

    # Refresh all players and get their positions
    all_players = list(player_repository.yield_players())
    
    # Update local player scatter
    player_scatter.setData([player_position.x], [player_position.y])

    # Separate players into enemy and friendly groups
    enemy_positions_x = []
    enemy_positions_y = []
    friendly_positions_x = []
    friendly_positions_y = []
    
    for player in all_players:
        if player.guid != local_player.guid and hasattr(player, 'position'):
            if local_player.is_enemy(player):
                enemy_positions_x.append(player.position.x)
                enemy_positions_y.append(player.position.y)
            else:
                friendly_positions_x.append(player.position.x)
                friendly_positions_y.append(player.position.y)
    
    # Update enemy and friendly scatter plots
    enemy_scatter.setData(enemy_positions_x, enemy_positions_y)
    friendly_scatter.setData(friendly_positions_x, friendly_positions_y)

    # Center the view on player position
    view_size = 250  # Adjust this value to control zoom level
    plot.setRange(xRange=[player_position.x - view_size, player_position.x + view_size],
                 yRange=[player_position.y - view_size, player_position.y + view_size])

# Set up timer for updates
timer = QtCore.QTimer()
timer.timeout.connect(update)
timer.start(100)  # Update every 100ms

# Start Qt event loop
if __name__ == '__main__':
    pg.exec()
```
